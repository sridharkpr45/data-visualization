```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
data= pd.read_csv("student.csv")
```

```python
data.head()
```

```python
df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x="grade", y="G3", data=df)
plt.title("Box Plot of Grade vs. G3 Score")
plt.xlabel("Grade")
plt.ylabel("G3 Score")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x="grade", y="G3", data=df)
plt.title("Violin Plot of Grade vs. G3 Score")
plt.xlabel("Grade")
plt.ylabel("G3 Score")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x="attendence", y="G3", data=df)
plt.title("Scatter Plot of Attendance vs. G3 Score")
plt.xlabel("Attendance")
plt.ylabel("G3 Score")
plt.show()
```
